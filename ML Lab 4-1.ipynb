```python
#df['Life expectancy'] = df['Life expectancy'].fillna(df['Life expectancy'].mean())
#df['Adult Mortality'] = df['Adult Mortality'].fillna(df['Adult Mortality'].mean())
df_new = df.fillna(df.mean(numeric_only = True))
df_new
```

```python
df_new.columns = df_new.columns.str.strip()
```

```python
df_new.columns = (
    df_new.columns
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
)

clean_df = df_new.drop(
    ['Adult Mortality', 'infant deaths', 'Measles',
     'under-five deaths', 'HIV/AIDS', 'Population',
     'thinness 1-19 years', 'thinness 5-9 years'],
    axis=1
)

clean_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
y = clean_df['Life expectancy']
x = clean_df.drop(['Life expectancy'], axis = 1)
x = pd.get_dummies(x, drop_first = True)

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 10)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
#y_pred
```

```python
#Evaluation
r2 = r2_score(y_test, y_pred)
print('R2 Score:', r2)
```

```python
#How much error is there in predicted
mse = mean_squared_error(y_test,y_pred)
print('Mean Squared Error: ',mse)
```

```python
#!pip install xgboost
#XGBoost Regression for the same dataset
from xgboost import XGBRegressor

xbr_model = XGBRegressor()
```

```python
xbr_model.fit(x_train,y_train)
```

```python
y_pred_xbr = xbr_model.predict(x_test)
#y_pred_xbr
```

```python
r2_xbr = r2_score(y_test,y_pred_xbr)
print("XGBoost R2 = ", r2_xbr)
```

Linear Regression was used as a baseline model to understand the relationship between health and socio-economic factors and life expectancy, but its performance was limited due to linear assumptions. XGBoost Regressor performed better by capturing non-linear relationships and complex feature interactions, resulting in higher prediction accuracy and lower error values.
